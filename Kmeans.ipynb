## DATA UNDERSTANDING

#Data Collection

```python
import pandas as pd
data = pd.read_csv('amazon.csv')
```

## Data Preparation

Membersihkan kolom actual_price dan rating_count dengan menghapus simbol non-numerik dan mengonversi ke float

```python
data['actual_price'] = data['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
data['rating_count'] = data['rating_count'].str.replace(',', '').astype(float)
```

```python
data.info()
```

Membersihkan data pada rating_count dengan menghapus missing value pada kolom tersebut

```python
missing_values = data['rating_count'].isnull().sum()
print(f"Jumlah missing value pada rating_count: {missing_values}")
```

```python
selected_data = data.dropna(subset=['rating_count'])
missing_values_after = selected_data['rating_count'].isnull().sum()
print(f"Jumlah missing value pada rating_count setelah penghapusan: {missing_values_after}")
```

```python
selected_data.info()
```

Memilih variabel yang akan digunakan dalam clustering yaitu: product_id, actual_price, rating_count

```python
selected_data = data[['product_id', 'actual_price', 'rating_count']].dropna()
```

Melakukan normalisasi data numerik pada kolom actual_price dan rating_count

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
selected_data[['actual_price', 'rating_count']] = scaler.fit_transform(selected_data[['actual_price', 'rating_count']])
```

## MODELING DAN EVALUATION

Menentukan Jumlah K Optimal dengan (Elbow Method)

```python
from sklearn.cluster import KMeans
inertia = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(selected_data[['actual_price', 'rating_count']])
    inertia.append(kmeans.inertia_)
```

Memvisualisasikan Elbow Method

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()
```

Membangun Model KMeans dengan nilai optimal

```python
define_k = 3  
kmeans = KMeans(n_clusters=define_k, random_state=42)
selected_data['cluster'] = kmeans.fit_predict(selected_data[['actual_price', 'rating_count']])
```

Menghitung Silhouette score untuk mengevaluasi kualitas klasterisasi.

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(selected_data[['actual_price', 'rating_count']], selected_data['cluster'])
print(f'Silhouette Score: {silhouette_avg}')
```

Nilai rata-rata Silhouette Score (dalam kasus ini, sekitar 0.756), yang menunjukkan bahwa klasterisasi memiliki pemisahan yang cukup baik (nilai mendekati 1 menunjukkan hasil klaster yang lebih baik).

Reduksi Dimensi menggunakan PCA agar data dapat divisualisasikan dalam 2 dimensi.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
selected_data_pca = pca.fit_transform(selected_data[['actual_price', 'rating_count']])
selected_data['pca1'] = selected_data_pca[:, 0]
selected_data['pca2'] = selected_data_pca[:, 1]
```

Visualisasi Klaster produk dengan Centroid

```python
# Plot hasil klasterisasi dengan centroid
plt.figure(figsize=(10, 7))
colors = ['red', 'blue', 'green', 'orange']
for cluster in range(define_k):
    cluster_data = selected_data[selected_data['cluster'] == cluster]
    plt.scatter(cluster_data['pca1'], cluster_data['pca2'], label=f'Cluster {cluster}', color=colors[cluster], alpha=0.6)

# Plot centroid
centroids_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], s=200, c='black', marker='X', label='Centroids')

plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Clustering Visualization with PCA and Centroids')
plt.legend()
plt.grid(True)
plt.show()
```

 Visualisasi Distribusi Data Berdasarkan Klaster Dengan Histogram 

```python
plt.figure(figsize=(12, 8))

for cluster in range(define_k):
    cluster_data = selected_data[selected_data['cluster'] == cluster]
    plt.subplot(define_k, 1, cluster + 1)
    plt.hist(cluster_data[['actual_price', 'rating_count']], bins=10, alpha=0.7, label=['Actual Price', 'Rating Count'], color=['orange', 'blue'])
    plt.title(f'Cluster {cluster} Characteristics')
    plt.xlabel('Normalized Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.tight_layout()

plt.show()
```

Sebagian besar produk di Cluster 0 memiliki nilai actual price yang rendah, tetapi beberapa produk memiliki actual price tinggi dengan mayoritas ulasan rendah.

Cluster 1:

Rating count tersebar lebih luas, dengan beberapa produk yang memiliki jumlah ulasan yang sangat tinggi meskipun harganya murah.

Cluster 2:

Harga produk dalam cluster ini lebih bervariasi dibandingkan dengan cluster lainnya, meskipun mayoritas masih berada di rentang rendah hingga menengah. Dan Jumlah ulasan untuk cluster ini juga lebih tersebar dengan rating tinggi.
